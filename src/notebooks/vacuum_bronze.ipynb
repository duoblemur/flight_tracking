```python
TARGET_TABLE = "dbr_dev.skybound_bronze.flights_bronze"
RETENTION_HOURS = 1

result = spark.sql(f"""
    DELETE FROM {TARGET_TABLE}
    WHERE ingestion_timestamp < (now() - INTERVAL {RETENTION_HOURS} HOURS)
""")
print(f"DELETE completed for {TARGET_TABLE}")

spark.sql(f"VACUUM {TARGET_TABLE} RETAIN {RETENTION_HOURS} HOURS")
print(f"VACUUM completed for {TARGET_TABLE}")
```
