```python
%sql
SELECT
  f.icao24,
  f.callsign,
  a.registration,
  a.manufacturerName,
  a.model,
  a.operator,
  a.country AS aircraft_country,
  f.origin_country AS flight_origin,
  f.latitude,
  f.longitude,
  f.altitude,
  f.speed,
  f.heading,
  f.on_ground,
  f.last_seen
FROM dbr_dev.skybound_silver.flights_silver f
INNER JOIN dbr_dev.skybound_silver.aircrafts_silver a
  ON f.icao24 = a.icao24
LIMIT 25;
```
